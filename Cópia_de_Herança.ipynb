# Pilares da OO: Herança

> **Herança**: Permite criar uma classe a partir de outra já existente, sendo a classe derivada (subclasse) herdeira de todos os atributos e métodos da classe principal (superclasse).

## O que já conhecemos: classes independentes

Imagine o sistema de uma empresa que registra nome, CPF e salário de todos os seus funcionários. Os funcionários da gerência são os únicos que tem acesso ao sistema, e portanto login e senha.

No início, vamos criar as duas classes **sem herança**, como se fossem independentes.

````python
## classe 1: Funcionário é a super classe
class Funcionario:
    def __init__(self, nome, cpf, salario):
        self.nome = nome
        self.cpf = cpf
        self.salario = salario

    def imprimir_dados(self):
        print(f"Funcionário: {self.nome} ({self.cpf})")
        print(f"Salário: R$ {self.salario:.2f}")

## classe 2: Gerente
class Gerente(Funcionario): # é uma subclasse de funcionario
    def __init__(self, nome, cpf, salario, login, senha):
        super().__init__(nome, cpf, salario)
        self.login = login
        self.senha = senha

    def imprimir_dados(self):
        super().imprimir_dados()
        print(f"Login: {self.login}, senha: {self.senha}")


    def validar_login(self, login, senha):
        return login == self.login and senha == self.senha

## Manipulando objeto do tipo Funcionário
f1 = Funcionario('Julia', '010.212.323-45', 5430.79)
f1.imprimir_dados()
print('-'*30)

## Manipulando objeto do tipo Gerente
g1 = Gerente('Aline', '002.323.434-56', 10430.79, 'line', 'line123')
g1.imprimir_dados()
````

## Herança: relação "é um"

As duas classes possuem:

- atributos `nome`, `CPF`, `salario`
- método `imprimir_dados()`

Faz sentido para o problema afirmar que todo **`Gerente` é um `Funcionário`**. Essa relação é refletida no código pela herança.

Nessa formulação, um gerente é uma versão mais especializada de funcionário, ou seja:

- `Funcionario` → Superclasse
- `Gerente` → Subclasse

Vamos reescrever as classes fazendo `Gerente` herdar de `Funcionario`.


````python
## Superclasse: Funcionário
class Funcionario:
    def __init__(self, nome, cpf, salario):
        self.nome = nome
        self.cpf = cpf
        self.salario = salario

    def imprimir_dados(self):
        print(f"Funcionário: {self.nome} ({self.cpf})")
        print(f"Salário: R$ {self.salario:.2f}")

## Subclasse: Gerente
class Gerente(Funcionario): # nome da superclasse entre parênteses na definição
    def __init__(self, nome, cpf, salario, login, senha):
        ### inicializando atributos da superclasse
        super().__init__(nome, cpf, salario)

        ### inicializando meus próprios atributos
        self.login = login
        self.senha = senha

    def imprimir_dados(self):
        super().imprimir_dados() # chamando método da superclasse
        print(f"Login: {self.login}, senha: {self.senha}")

    def validar_login(self, login, senha):
        return login == self.login and senha == self.senha


## Manipulando objeto do tipo Funcionário
f1 = Funcionario('Julia', '010.212.323-45', 5430.79)
f1.imprimir_dados()
print('-'*30)

## Manipulando objeto do tipo Gerente
g1 = Gerente('Aline', '002.323.434-56', 10430.79, 'line', 'line123')
g1.imprimir_dados()
print('Tentativa de login: ', g1.validar_login('aline',  '123456')) # retorna False
````

### O que mudou?

- A definição da classe
```python
class Subclasse(Superclasse)`
```

- Não precisamos redefinir na subclasse os atributos `nome`, `CPF` e `salario`, herdamos da superclasse.
   - Também não precisamos redefinir o método `imprimir_dados()` a menos que ele tenha um comportamento específico. **Experimente deletá-lo da subclasse**.

- Usamos o comando `super()` para acessar elementos da superclasse, inclusive o cosntrutor

```python
super().__init__(nome, cpf salario)
```

## Encapsulamento: níveis de acesso/visibilidade

Agora vamos falar sobre algo importante: **quem pode acessar os atributos no caso da herança?**

Vamos relembrar as convenções de nomenclatura do Python:

* `var_name`: **público**. Variáeis declaradas sem nenhum underline no início podem ser publicamente acessadas.

* `_um_underline`: **protegido**. Indicador fraco de "uso interno". Por convenção


* `__dois_underlines`: **privado**. Ao nomear um atributo ou método de classe,
provoca a alteração de nomes (dentro da classe Exemplo,
__x torna-se _Exemplo__x).

Guia de estilos do Python: https://peps.python.org/pep-0008/#descriptive-naming-styles

````python
class Funcionario:
    def __init__(self, nome, cpf, salario):
        self._nome = nome # protegido (sub pode acessar)
        self._cpf = cpf   # protegido
        self._salario = salario # protegido

    def imprimir_dados(self):
        print(f"Funcionário: {self._nome} ({self._cpf})")
        print(f"Salário: R$ {self._salario:.2f}")


class Gerente(Funcionario):
    def __init__(self, nome, cpf, salario, login, senha):
        super().__init__(nome, cpf, salario)

        self.__login = login  # privado
        self.__senha = senha  # privado

    def boas_vindas(self,):
        # subclasses podem acessar atributos protegidos da super
        print(f'Olá {self._nome}')

    def validar_login(self, login, senha):
        return login == self.login and senha == self.senha

f1 = Funcionario('Julia', '010.212.323-45', 5430.79)
f1.imprimir_dados()

g1 = Gerente('Aline', '002.323.434-56', 10430.79, 'line', 'line123')
g1.boas_vindas()
````
